```python
from pycaret.regression import *
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/pycaret/pycaret/master/datasets/insurance.csv')
```

```python
r2 = setup(data, target = 'charges', session_id = 123,
           normalize = True,
           polynomial_features = True, trigonometry_features = True,
           feature_interaction=True, 
           bin_numeric_features= ['age', 'bmi'])
```

```python
lr = create_model('lr')
```

```python
plot_model(lr, plot='residuals')
```

```python
save_model(lr, model_name='deploy_some_model')
```
